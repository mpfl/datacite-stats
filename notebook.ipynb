# DataCite stats

```python
import requests
import json
```

```python
url = "https://api.test.datacite.org/providers?consortium-id=ardc&page[size]=1000"

headers = {"accept": "application/vnd.api+json"}

response = requests.get(url, headers=headers)
```

```python
providers = json.loads(response.text)

len(providers["data"])

total = 0

for provider in providers["data"]:
    print(provider["id"] + ": " + str(len(provider["relationships"]["clients"]["data"])))
    total += len(provider["relationships"]["clients"]["data"])

print(total)
```
